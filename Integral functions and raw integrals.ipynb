# This Notebook presents a method of verifying the validity of an antiderivative of a function.

## Plotting the function's raw integral

The function is first integrated using the definition of an integral, we calculate the incremental sums of the $f(a+\Delta x*i)*\Delta x$. We call that the raw integral of the function. We then plot this raw integral and proceed to calculate the antiderivative of the function using calculus methods. If the antiderivative was correctly calculated, then its function must produce a curve similar (if not exactly the same) to the one we get when plotting the raw integral. Recall that the raw integral is obtained using computational methods, and the antiderivative using calculus methods. The raw integral function and the anti derivative may at times appear shifted, but we can understand why this is so, and that the shift can be accounted for using the constant of integration.

```python
import math
import matplotlib.pyplot as plt    # visualization library
import numpy as np                 # math library for the (sum)
def f(x):                          # we define our function
    return math.log(x)
    
g = []                             # the integral values as function product with delx f(a+i*delx)*delx, when in g, they give a list that produce a shape like the function f
x = []
delx = 0.1
n = 0
a = 1
b = 100

for i in range(b):
    xval = a+n
    n = n+delx
    x.append(xval)

for i in range(b):
    A = delx*f(a+delx*i)
    g.append(A)
    
intr = []
for i in range(len(g)):
    A = np.sum(g[0:i])
    intr.append(A)
    
fig, ax = plt.subplots(figsize = (20,5))
ax.plot(x,[math.log(i) for i in x], color = 'r') #function part
ax.plot(x, intr, color = 'blue') # integral part
```

# Plotting the antiderivative function

```python
def H(x):
    return math.log(x)*x-x
funct = [H(i) for i in x]

fig, ax = plt.subplots(figsize = (20,7))
ax.plot(x, funct)
```
